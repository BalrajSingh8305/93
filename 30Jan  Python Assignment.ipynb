## Question 5

#### In Python while loop is used to run a block code until a given certain condition is met
""" The condition is evaluated at the starting of each iteration of the loop. If the condition is True, then the code inside the loop is executed. 
Once the code inside the loop has finished executing, then again the condition is reevaluated, and the process repeats until the condition becomes False."""

```python
## Example 
count = 1
while count <= 10: ## here it is given that after the value gretaed then 10 it is fail
    print(count) ## to store the true condtion 
    count += 1 ## increace the value in each iteration
```

## Question 6

```python
## example 1.)
rows = int(input("enter the number"))
i = 1
while i <= rows:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i += 1
```

```python
rows = int(input("enter the number"))
i = rows
while i >= 1:
    j = i
    while j >= 1:
        print("*", end="")
        j -= 1
    print()
    i -= 1
```
